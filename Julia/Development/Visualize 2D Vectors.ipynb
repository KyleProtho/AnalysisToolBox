# Create 2D Vector Plots with Julia

Sometimes, it may be useful to visualize 2-dimensional vectors as you transform and analyze your data. This is often a skipped step, but can help to visualize and simplify how certain conclusions and operations are being reached computationally.

## 0. Import packages

```julia
# import Pkg
# Pkg.add("LinearAlgebra")
# Pkg.add("Plots")
using LinearAlgebra
using DataFrames
using Plots
theme(:vibrant)
```

## 1. Create a 2d vector

```julia
v1 = [2 4]
```

## 2. Plot 1 vector

We will need to use the quiver function to plot a vector from the origin point (0, 0). Use the vector created to population the quiver function and create a plot.

```julia
V = [0 0; v1[[1]] v1[[2]]]
```

```julia

gr() # Set the backend to GR
p = plot(
    V,
    title = "Plot of v1",
    xlims = (maximum(v1) * -1 - 2, maximum(v1) + 2),
    ylims = (maximum(v1) * -1 - 2, maximum(v1) + 2),
    draw_arrow = true,
    grid = :xy,
    gridlinewidth = 1,
    minorgrid = true
)
scatter!(
    x = V[:, 1], 
    y = V[:, 2]
)
plot(p, layout = 1)
```
